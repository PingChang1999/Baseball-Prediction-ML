```python
# Creating a table with every game since 1990
df = pd.DataFrame()
for year in range(1990, 2023):
    file_name = 'game_data/gl' + str(year) + '.txt'
    df_temp = pd.read_csv(file_name, header=None)
    df_temp.columns = col_names
    df_temp['season'] = year
    df = pd.concat((df, df_temp))
```

```python
# Calculating more columns
df['run_diff'] = df['runs_h'] - df['runs_v'] # home - visitors
df['home_victory'] = (df['run_diff'] > 0).astype(int) # did home team win
df['runs_total'] = df['runs_h'].copy() + df['runs_v'].copy()
df['date_dblhead'] = (df['date'].astype(str) + df['dblheader_code'].astype(str)).astype(int)
```
